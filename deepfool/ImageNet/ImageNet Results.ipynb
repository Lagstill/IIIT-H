```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

## AlexNet

### alexnet = models.alexnet(pretrained=True)

```python
img = mpimg.imread('cock_alex.png')

fig = plt.figure(figsize=(25, 25))
ax=fig.add_subplot(1,2,1)
ax.set_xlabel('Original Image', fontsize=30)
plt.imshow(img)

img = mpimg.imread('macaw_alex.png')
ax=fig.add_subplot(1,2,2)
ax.set_xlabel('Deepfooled Image', fontsize=30)
    
plt.imshow(img)
plt.show()
```
